## Analyze Covid-19 data
### Reading and converting file excel to table SQL

```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

pd.options.display.max_rows=999

%load_ext sql
%sql sqlite://
```

```python
df_d= pd.read_excel('CovidDeaths.xlsx',sheet_name='Sheet1')
df_v= pd.read_excel('CovidVaccination.xlsx',sheet_name='Sheet1')
```

```python
df_d.to_sql('dea',con=engine,index=False)
df_v.to_sql('vac',con=engine,index=False)
```

## Exploring data

```python
result = engine.execute("SELECT *  \
                           FROM dea \
                          ORDER BY 3,4\
                        ")
pd.DataFrame(result,columns=df_d.columns)
```

```python
result = engine.execute("SELECT *  \
                           FROM vac \
                          ORDER BY 3,4\
                        ")
pd.DataFrame(result,columns=df_v.columns)
```

### Select columns that we are starting with

```python
result = engine.execute("\
                         SELECT Location, date, total_cases, new_cases, total_deaths, population \
                           FROM dea \
                          ORDER BY 1,2\
                        ")
pd.DataFrame(result,columns=['Location', 'date', 'total_cases', 'new_cases', 'total_deaths', 'population'])
```

```python
# Look at total deaths and total cases of United States
result = engine.execute("SELECT Location, date, total_cases, total_deaths, (total_deaths/total_cases)*100 \
                           FROM dea \
                          WHERE Location LIKE '%state%' \
                          ORDER BY 2\
                        ")
pd.DataFrame(result,columns=['Location', 'date', 'total_cases', 'total_deaths', 'DeathPercentage'])
```

```python
# Looking at total cases vs population
result = engine.execute("SELECT Location, date, total_cases, population, (total_cases/population)*100 \
                           FROM dea \
                          ORDER BY 1,2\
                        ")
pd.DataFrame(result,columns=['Location', 'date', 'total_cases', 'population', 'CasePercentage'])
```

```python
# Looking at Countries with highest infection rate campared to population
result = engine.execute("SELECT Location, MAX(total_cases),population, ROUND((total_cases/population)*100,2) \
                           FROM dea \
                          GROUP BY 1,3 \
                          ORDER BY 4 DESC\
                        ")
pd.DataFrame(result,columns=['Country','HighestInfection', 'population', 'PercentPopulationInfected'])
```

```python
# Looking at countries with highest death count per population
result = engine.execute("SELECT Location, MAX(total_deaths)\
                           FROM dea \
                          WHERE continent IS NOT NULL \
                          GROUP BY 1 \
                          ORDER BY 2 DESC\
                        ")
pd.DataFrame(result,columns=['Country','HighestDeathCount'])
```

```python
# Looking at highest death count following by continent
result = engine.execute("SELECT continent, MAX(total_deaths)\
                           FROM dea \
                          WHERE continent IS NOT NULL \
                          GROUP BY 1 \
                          ORDER BY 2 DESC\
                        ")
pd.DataFrame(result,columns=['Continent','HighestDeathCount'])
```

```python
# Global numbners by day
result = engine.execute("SELECT date, SUM(new_cases), SUM(new_deaths),SUM(new_deaths)/SUM(new_cases) \
                           FROM dea \
                          WHERE continent IS NOT NULL \
                          GROUP BY date \
                          ORDER BY 1,2\
                        ")
pd.DataFrame(result,columns=['date','new_cases','new_deaths','DeathPercentage'])
```

```python
# Joining two table to each others
result = engine.execute("SELECT * \
                           FROM dea \
                           JOIN vac ON dea.Location= vac.Location \
                                   AND dea.date= vac.date \
                          WHERE dea.continent IS NOT NULL \
                          ORDER BY 3,4\
                        ")
pd.DataFrame(result,columns=df_d.columns.append(df_v.columns))
```

```python
# Looking at total population vs vaccinations
result = engine.execute("SELECT dea.continent, dea.location, dea.date, dea.population, vac.new_vaccinations \
                           FROM dea \
                           JOIN vac ON dea.Location= vac.Location \
                                   AND dea.date= vac.date \
                          WHERE dea.continent IS NOT NULL \
                          ORDER BY 2,3\
                        ")
pd.DataFrame(result,columns=['continent','location','date','population','new_vaccinations'])
```

```python
# Showing Percentage of Population that has recieved at least one Covid Vaccine
result = engine.execute("\
                        SELECT dea.continent, dea.location, dea.date, dea.population, vac.new_vaccinations\
                             , SUM(vac.new_vaccinations) OVER (PARTITION BY dea.Location \
                                                                   ORDER BY dea.location, dea.Date)\
                          FROM dea \
                          JOIN vac ON dea.Location= vac.Location \
                                   AND dea.date= vac.date \
                        WHERE dea.continent IS NOT NULL\
                        ORDER BY 2,3\
                        ")
pd.DataFrame(result,columns=['Continent','Location','date','Population','new_vaccinations','rolling_new_vaccinations'])
```
